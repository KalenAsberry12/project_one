```python
import pandas as pd
from pathlib import Path
%matplotlib inline
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
political_instability = Path("../project_one/SFIv2018.csv")
pol_inst_df = pd.read_csv(political_instability)
pol_inst_df
```

```python
pol_inst_df.columns
```

```python
pol_inst_df_2 = pol_inst_df.drop(columns=['scode', 'region', 'effect', 'legit',
       'seceff', 'secleg', 'poleff', 'polleg', 'ecoeff', 'ecoleg', 'soceff',
       'socleg'])
pol_inst_df_2
```

```python
pol_inst_df_3 = pol_inst_df_2.pivot(index='country', columns='year', values='sfi').dropna()
pol_inst_df_3.shape
```

```python
num_countries = 161
```

```python
avg_pol_inst = pol_inst_df_3.sum()/num_countries
```

```python
avg_pol_inst_df = pd.DataFrame(avg_pol_inst, columns=['Average Global Instability'])
avg_pol_inst_df
```

```python
avg_pol_inst_df.plot()
```
